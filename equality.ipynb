```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


players = 100  ##number of players
money = 100    ##inital money
iterations = 1000


##for now:

        #each player can give money to himself
        #players can go into negative
```

```python
main = np.full(players, money)
```

```python

simulation_result = pd.DataFrame()
simulation_result[0] = main


for k in range(1, iterations):
    for i in range(players):

        winner = np.random.randint(players, size=1) 
        main[winner] = main[winner]+1
        main[i] = main[i]-1

        simulation_result[k] = main
    

        
```

```python

fig, ax = plt.subplots(figsize=(10, 7))
def draw_simulation(index):
    

    ax.clear()
    ax.hlines(money,0,players-1, lw=1, color='#444444')

        
    dff = simulation_result[index]
    ax.plot(dff, marker='o', lw=3, color='#444444')

    ax.set_ylim((0, money*2))   # set the xlim to left, right
    ax.xaxis.set_ticks_position('bottom')
    ax.tick_params(axis='x', labelsize=14)
    ax.tick_params(axis='y', labelsize=14)

    ax.margins(0, 0.01)
    ax.grid(which='major', axis='x', linestyle='-')
    #ax.grid(which='major', axis='y', linestyle='-')

    ax.set_axisbelow(True)
    ax.text(0, 1.05, 'Simulation #'+str(index),
            transform=ax.transAxes, size=24, weight=600, ha='left')
    ax.set_xlabel('Player', size=20, weight=600)
    ax.set_ylabel('Money', size=20, weight=600)
    
draw_simulation(999)
```

```python
import matplotlib.animation as animation
from IPython.display import HTML

fig, ax = plt.subplots(figsize=(10,7))
animator = animation.FuncAnimation(fig, draw_simulation, frames=range(iterations))
```
